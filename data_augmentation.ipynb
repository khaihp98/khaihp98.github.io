```
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from matplotlib.pyplot import imread, imshow, subplots, show
```

```
def plot(data_generator):
    """
    Plots 4 images generated by an object of the ImageDataGenerator class.
    """
    data_generator.fit(images)
    image_iterator = data_generator.flow(images)
    
    # Plot the images given by the iterator
    fig, rows = subplots(nrows=1, ncols=4, figsize=(18,18))
    for row in rows:
        row.imshow(image_iterator.next()[0].astype('int'))
        row.axis('off')
    show()
    
```

```
image = imread("cat.jpeg")

# Creating a dataset which contains just one image.
images = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))

imshow(images[0])
show
```

```
data_generator = ImageDataGenerator(rotation_range=90)
plot(data_generator)
```

```
width_shift_generator = ImageDataGenerator(width_shift_range=0.3)
plot(width_shift_generator)
```

```
height_shift_generator = ImageDataGenerator(height_shift_range=0.3)
plot(height_shift_generator)
```

```
brightness_generator = ImageDataGenerator(brightness_range=(0.1, 0.9))
plot(brightness_generator)
```

```
shear_intensity_generator = ImageDataGenerator(shear_range=45.0)
plot(shear_intensity_generator)
```

```
zoom_generator = ImageDataGenerator(zoom_range=[0.5, 1.5])
plot(zoom_generator)
```

```
channel_shift_generator = ImageDataGenerator(channel_shift_range=150.0)
plot(channel_shift_generator)
```

```
horizontal_flip_generator = ImageDataGenerator(horizontal_flip=True)
plot(horizontal_flip_generator)
```

```
vertical_flip_generator = ImageDataGenerator(vertical_flip=True)
plot(vertical_flip_generator)
```

```
data_generator_nearest = ImageDataGenerator(width_shift_range=0.3, fill_mode='nearest')
plot(data_generator_nearest)
```

```
data_generator_reflect = ImageDataGenerator(width_shift_range=0.3, fill_mode='reflect')
plot(data_generator_reflect)
```

```
data_generator_wrap = ImageDataGenerator(width_shift_range=0.3, fill_mode='wrap')
plot(data_generator_wrap)
```

```
data_generator_contrast = ImageDataGenerator(width_shift_range=0.3, 
                                    fill_mode='constant', 
                                    cval=190)
plot(data_generator_contrast)
```

```
data_gen = ImageDataGenerator(zca_whitening=True)
plot(data_gen)
```
